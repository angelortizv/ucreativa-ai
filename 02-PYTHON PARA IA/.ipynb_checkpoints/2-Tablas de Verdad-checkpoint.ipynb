## Tablas de verdad

Las tablas de verdad nos permiten visualizar los valores resultantes de las operaciones l√≥gicas para todas las combinaciones posibles de verdad de sus operandos.

```python
import itertools
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def truth_table():
    data = []
    for p, q in itertools.product([True, False], repeat=2):
        row = {
            'p': p, 'q': q,
            'AND (p ‚àß q)': p and q,
            'OR (p ‚à® q)': p or q,
            'XOR (p ‚äï q)': p != q,
            'NAND (¬¨(p ‚àß q))': not (p and q),
            'NOR (¬¨(p ‚à® q))': not (p or q),
            'IMPLICA (p ‚Üí q)': not p or q,
            'BICONDICIONAL (p ‚Üî q)': p == q
        }
        data.append(row)
    df = pd.DataFrame(data)
    print(df)

    # Visualizaci√≥n de la tabla de verdad
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.iloc[:, 2:].astype(int), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.xticks(rotation=45, ha='right')
    plt.title("Tabla de Verdad - Operadores L√≥gicos")
    plt.show()

truth_table()
```
