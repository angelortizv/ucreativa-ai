# VISUALIZACI칍N DE DATOS INTRO

### 游늵 VISUALIZACI칍N DE DATOS
La visualizaci칩n de datos es una herramienta fundamental en el an치lisis de datos, ya que permite representar la informaci칩n de manera gr치fica mediante el uso de gr치ficos, diagramas, mapas de calor, histogramas, l칤neas de tiempo, entre otros.

Aunque los datos pueden ser almacenados y manipulados mediante estructuras como los DataFrames de Pandas, muchas veces resulta dif칤cil identificar patrones, relaciones o tendencias solo observando n칰meros en tablas.

### 游댌 쯇or qu칠 es importante la visualizaci칩n de datos?
1. Identificaci칩n de patrones: La visualizaci칩n hace posible detectar comportamientos, distribuciones, anomal칤as o correlaciones entre variables que podr칤an pasar desapercibidos en una tabla num칠rica.
2. Facilita la toma de decisiones: Al convertir datos complejos en representaciones visuales comprensibles, se facilita la toma de decisiones basada en evidencia.
3. Comunicaci칩n efectiva: Los analistas de datos no solo deben encontrar respuestas, sino tambi칠n comunicar sus hallazgos de manera clara y convincente. Las visualizaciones ayudan a transmitir ideas a p칰blicos diversos, incluyendo personas sin experiencia t칠cnica.

### Tipos de Gr치ficos

- Gr치fico de l칤neas: Muestra el comportamiento de los datos en un periodo de tiempo. Es 칰til para observar tendencias o fluctuaciones.
- Diagrama de barras: Sirve para comparar categor칤as de datos o ver cambios en el tiempo. Es claro y f치cil de interpretar.
- Diagrama de pastel: Representa partes de un todo y muestra los porcentajes o fracciones de cada parte respecto al total.
- Gr치fico de dispersi칩n: Muestra la relaci칩n entre dos variables mediante puntos individuales. Ayuda a identificar correlaciones.
- Histogramas: Representan la distribuci칩n de frecuencias de una variable. Se usa com칰nmente para observar la forma de los datos (simetr칤a, concentraci칩n, etc.).

<h3>Gr치ficos Lineales</h3>

```python
valoresEjeX = ['XVII', 'XVIII', 'XIX', 'XX', 'XXI']
valoresEjeY = [0.2, 0.5, 1.1, 2.2, 7.7]
plt.plot(valoresEjeX, valoresEjeY)  ## Crea el grafico con la relacion de las 2 listas
plt.title("Poblaci칩n mundial en los 칰ltimos siglos")  ## T칤tulo que tendra la gr치fica
plt.xlabel("Siglo") # Etiqueta del eje X
plt.ylabel("Poblaci칩n en Billones") # Etiqueta del eje Y, Bill칩n Americano
plt.show() # Para mostrar la gr치fica
```

Al analizar la gr치fica podemos ver que el comportamiento cambia en el siglo XX, 쯈u칠 sucesos ocurrieron para generar este comportamiento? <br/>

<ul>
    <li>Guerras Mundiales</li>
    <li>Mayor periodos de paz</li>
    <li>Desarrollo Tecnol칩gico</li>
</ul>

```python
# Dibujar multiples lineas en una gr치fica
ValoresEjeYSimulados = [0.2, 0.5, 1.1, 2.2, 4.4]
plt.plot(valoresEjeX, valoresEjeY, "bx-", label="Real") # bx --> blue con puntos X - linea continua
plt.plot(valoresEjeX, ValoresEjeYSimulados, "go--", label="Simulaci칩n")  # go -> green con puntos O -- linea cortada
plt.title("Poblaci칩n mundial en los 칰ltimos siglos")  ## T칤tulo que tendra la gr치fica
plt.xlabel("Siglo") # Etiqueta del eje X
plt.ylabel("Poblaci칩n en Billones") # Etiqueta del eje Y
plt.legend() # Muestra las etiquetas en la esquina
plt.show() # Para mostrar la gr치fica
```

<h3>Diagrama de Pastel</h3>

```python
Cantidad = [18,28,12,30,24]
Frutas = ["Peras","Manzanas","Guayabas","Uvas","Cocos"]
colores = ["#EE66FF","#60D394","#AAF683","#FFD97D","#127885"]
plt.pie(Cantidad, labels=Frutas, autopct="%0.1f %%", colors=colores)
plt.axis("equal") # La gr치fica se ve simetrica, no achatada
plt.show()
```

```python
Cantidad = [18,28,12,30,24]
Frutas = ["Peras","Manzanas","Guayabas","Uvas","Cocos"]
colores = ["#EE66FF","#60D394","#AAF683","#FFD97D","#127885"]
desfase = (0.5, 0.2, 0, 0.1,0)
plt.pie(Cantidad, labels=Frutas, autopct="%0.1f %%", colors=colores, explode=desfase)
plt.axis("equal") # La gr치fica se ve simetrica, no achatada
plt.show()
```

<h3>Diagrama de Distribuci칩n</h3>

```python
from random import randint
import random
random.seed(10)  # permite que el resultado sea siempre el mismo
altura = [ randint(120, 190) for x in range(20) ]
peso = [ randint(45, 125) for x in range(20) ]
plt.scatter(peso, altura)
plt.title("Distribuci칩n Peso - Altura")
plt.xlabel("Peso (Kgs)")
plt.ylabel("Altura (centimetros)")
plt.show( )
```

<h3>Histograma</h3>

```python
import numpy as np
np.random.seed(5)
Edades = [ np.random.normal(loc=40, scale=10) for x in range(1000) ] # edades distribuidas alrededor de 40
plt.hist(Edades, bins=45) # bins es el numero de barras
plt.title("Edades por poblaci칩n")
plt.xlabel("Edad")
plt.ylabel("# de Personas")
plt.show( )
```

<h3>DIAGRAMA DE CAJA</h3>

Son diagramas utilizados para comparar un valor estadistico en tiempo o por categorias. Permite visualizar el valor m칤nimo, m치ximo, cuartiles con los valores de 25% y 75%, as칤 como la media.

```python
## Ejemplo promedio de vida
Mujeres, Hombres = [ 72, 68, 65, 77, 73, 71, 69 ] , [ 60, 65, 68, 61,63, 64 ]
Promedio_de_Vida = [ Mujeres, Hombres ]
plt.figure(figsize=(15, 8))
plt.boxplot(Promedio_de_Vida)
plt.xticks( [ 1, 2 ] , [ "Mujer" , "Hombre" ] )
plt.title("Promedio por Genero", fontsize=22)
plt.ylabel("Edad (a침os)", fontsize=14)
plt.xlabel("Genero", fontsize=14)
plt.show( )
```
