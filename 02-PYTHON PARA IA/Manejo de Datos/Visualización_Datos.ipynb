# VISUALIZACIÓN DE DATOS INTRO

### 📊 VISUALIZACIÓN DE DATOS
La visualización de datos es una herramienta fundamental en el análisis de datos, ya que permite representar la información de manera gráfica mediante el uso de gráficos, diagramas, mapas de calor, histogramas, líneas de tiempo, entre otros.

Aunque los datos pueden ser almacenados y manipulados mediante estructuras como los DataFrames de Pandas, muchas veces resulta difícil identificar patrones, relaciones o tendencias solo observando números en tablas.

### 🔍 ¿Por qué es importante la visualización de datos?
1. Identificación de patrones: La visualización hace posible detectar comportamientos, distribuciones, anomalías o correlaciones entre variables que podrían pasar desapercibidos en una tabla numérica.
2. Facilita la toma de decisiones: Al convertir datos complejos en representaciones visuales comprensibles, se facilita la toma de decisiones basada en evidencia.
3. Comunicación efectiva: Los analistas de datos no solo deben encontrar respuestas, sino también comunicar sus hallazgos de manera clara y convincente. Las visualizaciones ayudan a transmitir ideas a públicos diversos, incluyendo personas sin experiencia técnica.

### Tipos de Gráficos

- Gráfico de líneas: Muestra el comportamiento de los datos en un periodo de tiempo. Es útil para observar tendencias o fluctuaciones.
- Diagrama de barras: Sirve para comparar categorías de datos o ver cambios en el tiempo. Es claro y fácil de interpretar.
- Diagrama de pastel: Representa partes de un todo y muestra los porcentajes o fracciones de cada parte respecto al total.
- Gráfico de dispersión: Muestra la relación entre dos variables mediante puntos individuales. Ayuda a identificar correlaciones.
- Histogramas: Representan la distribución de frecuencias de una variable. Se usa comúnmente para observar la forma de los datos (simetría, concentración, etc.).

<h3>Gráficos Lineales</h3>

```python
valoresEjeX = ['XVII', 'XVIII', 'XIX', 'XX', 'XXI']
valoresEjeY = [0.2, 0.5, 1.1, 2.2, 7.7]
plt.plot(valoresEjeX, valoresEjeY)  ## Crea el grafico con la relacion de las 2 listas
plt.title("Población mundial en los últimos siglos")  ## Título que tendra la gráfica
plt.xlabel("Siglo") # Etiqueta del eje X
plt.ylabel("Población en Billones") # Etiqueta del eje Y, Billón Americano
plt.show() # Para mostrar la gráfica
```

Al analizar la gráfica podemos ver que el comportamiento cambia en el siglo XX, ¿Qué sucesos ocurrieron para generar este comportamiento? <br/>

<ul>
    <li>Guerras Mundiales</li>
    <li>Mayor periodos de paz</li>
    <li>Desarrollo Tecnológico</li>
</ul>

```python
# Dibujar multiples lineas en una gráfica
ValoresEjeYSimulados = [0.2, 0.5, 1.1, 2.2, 4.4]
plt.plot(valoresEjeX, valoresEjeY, "bx-", label="Real") # bx --> blue con puntos X - linea continua
plt.plot(valoresEjeX, ValoresEjeYSimulados, "go--", label="Simulación")  # go -> green con puntos O -- linea cortada
plt.title("Población mundial en los últimos siglos")  ## Título que tendra la gráfica
plt.xlabel("Siglo") # Etiqueta del eje X
plt.ylabel("Población en Billones") # Etiqueta del eje Y
plt.legend() # Muestra las etiquetas en la esquina
plt.show() # Para mostrar la gráfica
```

<h3>Diagrama de Barras</h3>

```python
categoriasGenerosMusicales = ['Salsa','Bachata','Cumbia','Reguetón','Rock','Tango','Samba' ]
cantidadPersonas = [5,3,4,7,10,6,9]
plt.bar(categoriasGenerosMusicales, cantidadPersonas)
plt.title("Categorías Preferidas Generos Musicales", fontsize=22, color="#AC501E")
plt.xlabel('Categorias Musicales', fontsize=14, color="#FF0000")
plt.ylabel('N° Personas', fontsize=14, color="#00FF00")
plt.figure(figsize=(15, 8))
plt.show()
```

<h3>Diagrama de Pastel</h3>

```python
Cantidad = [18,28,12,30,24]
Frutas = ["Peras","Manzanas","Guayabas","Uvas","Cocos"]
colores = ["#EE66FF","#60D394","#AAF683","#FFD97D","#127885"]
plt.pie(Cantidad, labels=Frutas, autopct="%0.1f %%", colors=colores)
plt.axis("equal") # La gráfica se ve simetrica, no achatada
plt.show()
```

```python
Cantidad = [18,28,12,30,24]
Frutas = ["Peras","Manzanas","Guayabas","Uvas","Cocos"]
colores = ["#EE66FF","#60D394","#AAF683","#FFD97D","#127885"]
desfase = (0.5, 0.2, 0, 0.1,0)
plt.pie(Cantidad, labels=Frutas, autopct="%0.1f %%", colors=colores, explode=desfase)
plt.axis("equal") # La gráfica se ve simetrica, no achatada
plt.show()
```

<h3>Diagrama de Distribución</h3>

```python
from random import randint
import random
random.seed(10)  # permite que el resultado sea siempre el mismo
altura = [ randint(120, 190) for x in range(20) ]
peso = [ randint(45, 125) for x in range(20) ]
plt.scatter(peso, altura)
plt.title("Distribución Peso - Altura")
plt.xlabel("Peso (Kgs)")
plt.ylabel("Altura (centimetros)")
plt.show( )
```

<h3>Histograma</h3>

```python
import numpy as np
np.random.seed(5)
Edades = [ np.random.normal(loc=40, scale=10) for x in range(1000) ] # edades distribuidas alrededor de 40
plt.hist(Edades, bins=45) # bins es el numero de barras
plt.title("Edades por población")
plt.xlabel("Edad")
plt.ylabel("# de Personas")
plt.show( )
```

<h3>DIAGRAMA DE CAJA</h3>

Son diagramas utilizados para comparar un valor estadistico en tiempo o por categorias. Permite visualizar el valor mínimo, máximo, cuartiles con los valores de 25% y 75%, así como la media.

```python
## Ejemplo promedio de vida
Mujeres, Hombres = [ 72, 68, 65, 77, 73, 71, 69 ] , [ 60, 65, 68, 61,63, 64 ]
Promedio_de_Vida = [ Mujeres, Hombres ]
plt.figure(figsize=(15, 8))
plt.boxplot(Promedio_de_Vida)
plt.xticks( [ 1, 2 ] , [ "Mujer" , "Hombre" ] )
plt.title("Promedio por Genero", fontsize=22)
plt.ylabel("Edad (años)", fontsize=14)
plt.xlabel("Genero", fontsize=14)
plt.show( )
```

<h3>Mapa de Calor</h3>

```python
import pandas as pd
historico_ventas = pd.DataFrame(
    np.random.randint(1, 100, 120).reshape(12, 10),
    columns = np.arange(2009, 2019),
    index = ["ene", "feb", "mar", "abr", "may", "jun", "jul", "ago", "sep", "oct", "nov", "dic"]
)
historico_ventas
```

```python
ax = sns.heatmap(historico_ventas, annot = True);  # El parámetro annot permite ver los números
```

Los valores más altos tienen a color blanco, aqui vemos el caso de que en Julio de 2013 se tuvo la mayor cantidad de ventas
