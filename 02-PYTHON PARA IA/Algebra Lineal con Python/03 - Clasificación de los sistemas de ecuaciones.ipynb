#  **SISTEMAS DE ECUACIONES CON DOS ECUACIONES Y DOS INCOGNITAS**

#  **Sistema compatible determinado**


$$\left\{
\begin{array}{rcl}
     4x+10y & = & 10
  \\ x+y & = & 4
\end{array}
\right.$$

###   **Soluci√≥n usando funci√≥n solve**

```python
Ecu1 = Eq(x+y,4)
Ecu2 = Eq(4*x+2*y,10)
solve([Ecu1,Ecu2])
```

###  **Representaci√≥n gr√°fica**


```python
plot2D(Ecu1,Ecu2,{x:1, y:3})
```

###   **Soluci√≥n con el m√©todo de Gauss-Jordan**

```python
A = Matrix([[4,2,10],[1,1,4]])
A
```

```python
A.rref(pivots=False)
```

#   **SISTEMAS DE ECUACIONES CON TRES ECUACIONES Y TRES INCOGNITAS**

#   **Sistema compatible determinado**


$$\left\{
\begin{array}{rcl}
     x+y-z & = & 2
  \\ 2x-y+z & = & 1
  \\ x-y+2z & = & 2
\end{array}
\right.$$


###   **Soluci√≥n usando funci√≥n solve**

```python
Ecu1 = Eq(x+y-z,2)
Ecu2 = Eq(2*x-y+z,1)
Ecu3 = Eq(x-y+2*z,2)
solve([Ecu1,Ecu2,Ecu3])
```

###   **Representaci√≥n gr√°fica**

```python
plot3D(Ecu1,Ecu2,Ecu3,{x:1, y:3, z:2})
```

###   **Soluci√≥n con el m√©todo de Gauss-Jordan**

```python
A = Matrix([[1,1,-1,2],[2,-1,1,1],[1,-1,2,2]])
A
```

```python
A.rref(pivots=False)
```

#   **Sistema compatible indeterminado**


$$\left\{
\begin{array}{rcl}
     x+y+3z & = & 2
  \\ 2x+3y+4z & = & 1
  \\ -2x-y-8z & = & -7
\end{array}
\right.$$


###   **Soluci√≥n usando funci√≥n solve**

```python
Ecu1 = Eq(x+y+3*z,2)
Ecu2 = Eq(2*x+3*y+4*z,1)
Ecu3 = Eq(-2*x-y-8*z,-7)
solve([Ecu1,Ecu2,Ecu3])
```

###   **Representaci√≥n gr√°fica**

```python
plot3D(Ecu1,Ecu2,Ecu3,{x:5-5*t, y:2*t-3,z:t,t:10})
```

###  **Soluci√≥n con el m√©todo de Gauss-Jordan**

```python
A = Matrix([[1,1,3,2],[2,3,4,1],[-2,-1,-8,-7]])
A
```

```python
A.rref(pivots=False)
```

#   **Sistema incompatible**


$$\left\{
\begin{array}{rcl}
     x+2y+4z & = & 3
  \\ 2x+4y+8z & = & 1
  \\ -y+z & = & -2
\end{array}
\right.$$


### **Soluci√≥n usando funci√≥n solve**

```python
Ecu1 = Eq(x+2*y+4*z,3)
Ecu2 = Eq(2*x+4*y+8*z,1)
Ecu3 = Eq(-y+z,-2)
solve([Ecu1,Ecu2,Ecu3])
```

###   **Representaci√≥n gr√°fica**

```python
plot3D(Ecu1,Ecu2,Ecu3)
```

###  **Soluci√≥n con el m√©todo de Gauss-Jordan**

```python
A = Matrix([[1,2,4,3],[2,4,8,1],[0,-1,1,-2]])
A
```

```python
A.rref(pivots=False)
```
