## ğŸ¯ Objetivo:
Simular dos bots que ejecutan tareas repetitivas al mismo tiempo, como si fueran procesos independientes.

```python
import threading
import time
import random
```

```python
def bot1():
    while True:
        time.sleep(random.randint(2, 4))
        print("ğŸ¤– Bot 1: Revisando correos...")

def bot2():
    while True:
        time.sleep(random.randint(3, 5))
        print("ğŸ¤– Bot 2: Generando reporte diario...")
```

```python
# Creamos los hilos (threads)
t1 = threading.Thread(target=bot1, daemon=True)
t2 = threading.Thread(target=bot2, daemon=True)

# Iniciamos los bots
t1.start()
t2.start()

# Monitor principal
for i in range(15):
    print(f"ğŸ§­ Monitor principal: ciclo {i+1}")
    time.sleep(1)

print("âœ… Sistema de bots terminado.")
```
