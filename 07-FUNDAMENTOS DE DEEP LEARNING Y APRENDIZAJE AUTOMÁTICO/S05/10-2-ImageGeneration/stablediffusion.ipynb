## 2. Importación de Librerías

```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

## 3. Introducción a la Generación de Imágenes

Los modelos como Stable Diffusion y DALL·E generan imágenes a partir de descripciones textuales. 
Estos modelos combinan técnicas de procesamiento del lenguaje natural y visión por computadora.


## 4. Cargar un Modelo Preentrenado

```python
# En este ejemplo, usaremos Stable Diffusion.

# Establecemos el modelo y verificamos si tenemos una GPU disponible.
device = "cuda" if torch.cuda.is_available() else "cpu"

# Cargamos el pipeline de Stable Diffusion.
pipeline = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipeline = pipeline.to(device)
```

## 5. Generación de una Imagen a partir de Texto

```python
# Creamos una descripción textual (prompt).
prompt = "A futuristic cityscape at sunset, with flying cars and neon lights"

# Generamos la imagen.
image = pipeline(prompt).images[0]

# Mostramos la imagen generada.
plt.imshow(image)
plt.axis("off")
plt.title("Imagen Generada a partir de Texto")
plt.show()
```

## 6. Exploración de Prompts

```python
# Probamos diferentes descripciones para observar cómo varía la salida.
def generate_and_show(prompt):
    print(f"Prompt: {prompt}")
    image = pipeline(prompt).images[0]
    plt.imshow(image)
    plt.axis("off")
    plt.title("Imagen Generada")
    plt.show()

# Prompts adicionales.
prompts = [
    "A fantasy forest with glowing mushrooms and a magical river",
    "A robotic dog playing in a park",
    "An ancient castle on a hill during a thunderstorm"
]

# Generamos y mostramos imágenes para cada prompt.
for p in prompts:
    generate_and_show(p)
```

## 7. Ajustes en la Generación
Los modelos permiten modificar parámetros como la **guidance scale** (grado de control) y **número de pasos**.


```python
# Prompt con ajustes personalizados.
prompt = "A hyperrealistic portrait of a cyborg with intricate mechanical details"
image = pipeline(prompt, guidance_scale=7.5, num_inference_steps=50).images[0]

# Mostramos la imagen generada con los ajustes.
plt.imshow(image)
plt.axis("off")
plt.title("Imagen con Ajustes Personalizados")
plt.show()
```
